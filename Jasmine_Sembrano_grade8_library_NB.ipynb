```python
import json
import requests

url = "https://raw.githubusercontent.com/AndreaJulieFSembrano/Jasmine-Sembrano-grade8-library-Repo/refs/heads/main/books.json"
response = requests.get(url)
books = response.json()
print(books)
```

```python
import json
import requests
url = "https://raw.githubusercontent.com/AndreaJulieFSembrano/Jasmine-Sembrano-grade8-library-Repo/refs/heads/main/firebase_key.json"
response = requests.get(url)
firebase_key = response.json()
print(firebase_key)
```

```python
import firebase_admin
from firebase_admin import credentials, db
cred = credentials.Certificate(firebase_key)
# Initialize the app with your database URL
firebase_admin.initialize_app(cred, {
 "databaseURL": "https://jasmine-sembrano-library-db-default-rtdb.asia-southeast1.firebasedatabase.app/"
})
print("Firebase connected successfully!")
```

```python
total = 0

for b in books["books"]:
    total += b["details"]["copies"]

print("Total number of books in library:", total)
```

```python
for b in books["books"]:
    print("ID:", b["id"])
    print("Title:", b["title"])
    print("Author:", b["author"])
    print("Copies:", b["details"]["copies"])
    print("Available:", b["details"]["available"])
    print("-" * 30)
```
